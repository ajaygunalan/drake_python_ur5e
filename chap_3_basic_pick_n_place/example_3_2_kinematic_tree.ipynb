```python
import pydot
from IPython.display import SVG, display
from pydrake.all import MultibodyPlant, Parser

def kinematic_tree_example():
    plant = MultibodyPlant(time_step=0.0)
    parser = Parser(plant)
    
    # Add package map for ur_description (official Universal Robots ROS2 package)
    parser.package_map().Add(
        'ur_description', 
        '/home/ajay/drake/Universal_Robots_ROS2_Description'
    )
    
    # Load official UR5e URDF (converted to Drake-compatible format)
    parser.AddModels('/home/ajay/drake/Universal_Robots_ROS2_Description/ur5e_drake.urdf')
    plant.Finalize()

    print(f"UR5e loaded: {plant.num_bodies()} bodies, {plant.num_joints()} joints")
    
    display(
        SVG(
            pydot.graph_from_dot_data(plant.GetTopologyGraphvizString())[0].create_svg()
        )
    )


kinematic_tree_example()
```
